## Creating Plots

```python
# Scatter plot

fig, ax = plt.subplots()
ax.scatter(x = data["Date"][:1000], # visualise the first 1000 values
           y= data["AQI"][:1000])
ax.set_xlabel("Date")
ax.set_ylabel("AQI")
ax.set_title("First 1000 datapoints plotted")
plt.show()
```

```python
# Histogram
data.AQI.plot.hist(xlabel="AQI")
```

#### Inference
We can see the most common AQI levels are in the 90-100 range, signifying reasonable AQI Metrics

```python
# Plotting the Yearly median AQI level

data.groupby(["Year"])["AQI"].median().plot()
plt.xlabel("Year")
plt.ylabel("Median AQI Level")
plt.show()
```

#### Inference
* 2020 had the lowest aqi levels due to the COVID Lockdown
* 2023 has seen an extreme increase in pollution levels as companies started enforcing Work from office policies

```python
# Boxplot to identify outlier data

plt.figure(figsize=(10,6))
sns.boxplot(x="Month", y="AQI", data=data)
plt.title("Distribution of AQI across each Month")
plt.ylabel("AQI")
plt.xlabel("Month")
plt.show()
```

#### Inference
* We can clearly see monsoon months have lower AQI Levels (outliers are also more concentrated)
* While winter months have higher Levels on an average

```python
# Creating and plotting correlation matrix

correlation_matrix = data.corr(numeric_only=True)
plt.figure(figsize=(8,5))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="crest")
```

#### Inference
We will be focusing on high correlation to AQI
* Minimum temprature has high negative correlation to AQI
* Weather code also signifies helps identify AQI levels (since a higher weather code signifies more rain and lower AQI Levels)
* Cloud coverage and wind speed also help identify AQI while wind direction has lower negative correlation
